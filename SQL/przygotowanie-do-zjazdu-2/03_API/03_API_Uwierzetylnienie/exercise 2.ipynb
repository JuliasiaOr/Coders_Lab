```python
API_KEY = 'af3f4cdb82efc648c561c7ee6ed27c13'
URL = 'https://api.openweathermap.org/data/2.5/weather'
```

```python
def get_weather(city, country=""): 
    q = ",".join([city, country])    
    params = {'q': q,
             'appid': API_KEY}    
    res = requests.get(URL, params=params)
    data = res.json()    
    if data.get('cod') == '404':
        return "error"
    else:
        return data.get('weather')[0].get('main')
```
