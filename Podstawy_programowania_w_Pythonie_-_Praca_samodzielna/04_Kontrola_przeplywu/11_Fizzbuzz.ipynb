### Kontrola przepływu > Fizzbuzz

Dla liczba z zakresu od 1 do 100 (włącznie), napisz program, który:
* dla liczby podzielnej przez 3 wypisze `Fizz` (przykładowo dla 6 ma zostać wypisane `Fizz`),
* dla liczby podzielnej przez 5 wypis `Buzz` (przykładowo dla 10 ma zostać wypisane `Buzz`),
* dla liczby podzielnej przez 3 i 5 wypisze `FizzBuzz` (przykładowo dla liczby 15 ma się wypisać **tylko** `FizzBuzz`),
* w przeciwnym wypadku program wypisze na ekran liczbę (przykładowo dla 4 ma zostać wypisane `4`).

```python
for i in range (0, 99):
    i += 1
    if i % 3 == 0:
        print(i, "Fizz")
    if i % 5 ==0:
        print(i, "Buzz")
    if i % 3 == 0 and i % 5 ==0:
        print(i, "FizzBuzz")
    else:
        print(i)
    

```
